```python
# Cell 1: Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cell 2: Load Data
results_df = pd.read_csv('../data/international_matches/results.csv')
print(results_df.info())
```

```python
# Cell 3: Visualize Historical Goal Distributions
plt.figure(figsize=(10, 5))
sns.histplot(results_df['home_score'] + results_df['away_score'], bins=20, kde=True)
plt.title('Distribution of Total Goals per Match')
plt.xlabel('Total Goals')
plt.show()
```

```python
# Cell 4: Win Rate by Home Advantage
home_wins = (results_df['home_score'] > results_df['away_score']).mean()
print(f"Historical Home Win Percentage: {home_wins*100:.2f}%")
```
