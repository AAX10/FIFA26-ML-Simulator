```python
import pandas as pd
import numpy as np
import joblib
import random
from collections import Counter
from pathlib import Path
import os

BASE_DIR = Path.cwd().parent

MODELS_DIR = BASE_DIR / "models"
DATA_DIR = BASE_DIR / "data"

clf = joblib.load(MODELS_DIR / "classifier.pkl")
reg_home = joblib.load(MODELS_DIR / "regressor_home.pkl")
reg_away = joblib.load(MODELS_DIR / "regressor_away.pkl")

team_profiles = pd.read_csv(DATA_DIR / "team_profiles.csv").set_index("Team")
fixtures_df = pd.read_csv(DATA_DIR / "world_cup_2026_fixtures.csv")

fixtures_df["Group"] = fixtures_df["Group"].astype(str).str.strip()
fixtures_df["Home"] = fixtures_df["Home"].astype(str).str.strip()
fixtures_df["Away"] = fixtures_df["Away"].astype(str).str.strip()

groups = pd.concat([
    fixtures_df[["Group", "Home"]].rename(columns={"Home": "Team"}),
    fixtures_df[["Group", "Away"]].rename(columns={"Away": "Team"})
]).drop_duplicates().sort_values(["Group", "Team"])

# --- STEP 1: Core Prediction Logic ---
def predict_match(home_team, away_team):
    if home_team not in team_profiles.index or away_team not in team_profiles.index:
        return None

    home = team_profiles.loc[home_team]
    away = team_profiles.loc[away_team]

    X = pd.DataFrame([{
        "Elo_Delta": home["Elo"] - away["Elo"],
        "Rank_Delta": away["Rank"] - home["Rank"],
        "Points_Delta": home["Points"] - away["Points"],
        "OVR_11_Delta": home["OVR11"] - away["OVR11"],
        "Value_Delta": np.log1p(home["Value"]) - np.log1p(away["Value"]),
        "WC_Pedigree_Delta": home["WC_Titles"] - away["WC_Titles"],
        "Home_Form_Pts": home["Form_Pts"],
        "Away_Form_Pts": away["Form_Pts"],
        "Form_Pts_Delta": home["Form_Pts"] - away["Form_Pts"],
        "Form_GS_Delta": home["Form_GS"] - away["Form_GS"],
        "Form_GC_Delta": home["Form_GC"] - away["Form_GC"],
        "Goal_Diff_Form": (home["Form_GS"] - away["Form_GS"]) - (home["Form_GC"] - away["Form_GC"]),
        "Weighted_Form_Delta": home["Weighted_Form"] - away["Weighted_Form"],
        "Form5_Delta": home["Form5"] - away["Form5"],
        "Form10_Delta": home["Form10"] - away["Form10"],
        "Form20_Delta": home["Form20"] - away["Form20"],
        "Momentum_Delta": home["Momentum"] - away["Momentum"]
    }])

    # Guarantee the feature order exactly matches what the model was trained on
    X = X[clf.feature_names_in_]

    # Probabilities and Base Scores
    probs = clf.predict_proba(X)[0]
    draw = probs[1] * 0.6
    remaining = 1 - draw
    win_total = probs[0] + probs[2]

    away_win = remaining * (probs[0] / win_total) if win_total > 0 else 0
    home_win = remaining * (probs[2] / win_total) if win_total > 0 else 0

    home_xg = max(
        0.1,
        float(reg_home.predict(X)[0])
    )
    
    away_xg = max(
        0.1,
        float(reg_away.predict(X)[0])
    )
    
    probs = clf.predict_proba(X)[0]
    
    home_win_prob = float(probs[2])
    draw_prob = float(probs[1])
    away_win_prob = float(probs[0])
    
    strength = home_win_prob - away_win_prob
    
    home_xg *= (1 + 0.30 * strength)
    away_xg *= (1 - 0.30 * strength)
    
    home_xg = max(home_xg, 0.1)
    away_xg = max(away_xg, 0.1)

    penalty_home = float(home["Penalty_Rating"])
    penalty_away = float(away["Penalty_Rating"])
    total = penalty_home + penalty_away
    if total == 0: total = 1

    home_penalty_prob = penalty_home / total
    away_penalty_prob = penalty_away / total
    penalty_winner = home_team if home_penalty_prob > away_penalty_prob else away_team

    return {
        "probabilities": {"home_win": float(home_win), "draw": float(draw), "away_win": float(away_win)},
        "predicted_score": {"home_goals": int(round(home_goals)), "away_goals": int(round(away_goals))},
        "knockout_prediction": {"winner": penalty_winner}
    }


# --- STEP 2: The Score Generator ---
def generate_score(result, deterministic=False):
    probs = result["probabilities"]

    if deterministic:
        # Maximum Likelihood branch for the final exported bracket
        max_prob = max(probs["home_win"], probs["draw"], probs["away_win"])
        if max_prob == probs["home_win"]: outcome = "H"
        elif max_prob == probs["away_win"]: outcome = "A"
        else: outcome = "D"
    else:
        # Monte Carlo random chance branch (normalized to prevent float errors)
        p_list = np.array([probs["home_win"], probs["draw"], probs["away_win"]])
        p_list /= p_list.sum()
        outcome = np.random.choice(["H", "D", "A"], p=p_list)

    hg = result["predicted_score"]["home_goals"]
    ag = result["predicted_score"]["away_goals"]

    # Introduce realistic variance during Monte Carlo simulations
    if not deterministic:
        hg = max(0, hg + np.random.randint(-1, 2))
        ag = max(0, ag + np.random.randint(-1, 2))

    # Enforce logical consistency between W/D/L outcome and goals
    if outcome == "H":
        if hg <= ag:
            hg = ag + 1
    elif outcome == "A":
        if ag <= hg:
            ag = hg + 1
    else:
        hg = ag

    return outcome, hg, ag


# --- STEP 3: Update simulate_group ---
def simulate_group(group, teams, all_matches, deterministic=False):
    table = pd.DataFrame(0, index=teams, columns=["P", "W", "D", "L", "Pts", "GF", "GA", "GD"], dtype=int)
    group_fixtures = fixtures_df[fixtures_df["Group"] == group].sort_values("Matchday")

    for _, match in group_fixtures.iterrows():
        home, away = match["Home"], match["Away"]
        result = predict_match(home, away)
        if result is None: continue

        # Replace random.choices with generate_score
        outcome, hg, ag = generate_score(result, deterministic)
        probs = result["probabilities"]

        table.loc[home, "P"] += 1
        table.loc[away, "P"] += 1
        table.loc[home, "GF"] += hg
        table.loc[home, "GA"] += ag
        table.loc[away, "GF"] += ag
        table.loc[away, "GA"] += hg

        if outcome == "H":
            table.loc[home, "Pts"] += 3
            table.loc[home, "W"] += 1
            table.loc[away, "L"] += 1
            winner_team = home
        elif outcome == "A":
            table.loc[away, "Pts"] += 3
            table.loc[away, "W"] += 1
            table.loc[home, "L"] += 1
            winner_team = away
        else:
            table.loc[home, "Pts"] += 1
            table.loc[away, "Pts"] += 1
            table.loc[home, "D"] += 1
            table.loc[away, "D"] += 1
            winner_team = "Draw"

        # STEP 5: Store Flattened Match Scores
        if deterministic and all_matches is not None:
            all_matches.append({
                "Stage": "Group",
                "Group": group,
                "Home": home,
                "Away": away,
                "HomeGoals": hg,
                "AwayGoals": ag,
                "Winner": winner_team,
                "HomeWinProb": probs["home_win"],
                "DrawProb": probs["draw"],
                "AwayWinProb": probs["away_win"]
            })

    table["GD"] = table["GF"] - table["GA"]
    table = table.sort_values(["Pts", "GD", "GF"], ascending=False)
    return table


# --- STEP 4: Update simulate_knockout ---
def simulate_knockout(team1, team2, stage, all_matches, deterministic=False):
    if team1 is None or team2 is None: return team1 if team1 else team2
    result = predict_match(team1, team2)
    if result is None: return team1

    # Replace random.choices with generate_score
    outcome, hg, ag = generate_score(result, deterministic)
    probs = result["probabilities"]
    
    if outcome == "H":
        winner = team1
    elif outcome == "A":
        winner = team2
    else:
        winner = result["knockout_prediction"]["winner"]

    # STEP 5: Store Flattened Match Scores
    if deterministic and all_matches is not None:
        all_matches.append({
            "Stage": stage,
            "Home": team1,
            "Away": team2,
            "HomeGoals": hg,
            "AwayGoals": ag,
            "Winner": winner,
            "HomeWinProb": probs["home_win"],
            "DrawProb": probs["draw"],
            "AwayWinProb": probs["away_win"]
        })

    return winner


def play_round(teams, stage, all_matches, deterministic=False):
    winners = []
    for i in range(0, len(teams), 2):
        winner = simulate_knockout(teams[i], teams[i + 1], stage, all_matches, deterministic)
        winners.append(winner)
    return winners


def simulate_world_cup(deterministic=False):
    all_matches = [] if deterministic else None
    group_winners, group_runners, all_group_tables = {}, {}, {}

    for group in sorted(groups["Group"].unique()):
        teams = groups[groups["Group"] == group]["Team"].tolist()
        table = simulate_group(group, teams, all_matches, deterministic)
        # STEP 6: Store all_group_tables
        all_group_tables[group] = table
        group_winners[group] = table.index[0]
        group_runners[group] = table.index[1]

    third_place = []
    for group, table in all_group_tables.items():
        if len(table) > 2:
            third_place.append({"Group": group, "Team": table.index[2], "Pts": table.iloc[2]["Pts"], "GD": table.iloc[2]["GD"], "GF": table.iloc[2]["GF"]})

    best_8 = pd.DataFrame(third_place).sort_values(["Pts", "GD", "GF"], ascending=False).head(8)
    best_third_teams = best_8["Team"].tolist()

    round_of_32 = [
        (group_winners.get("A"), best_third_teams[7] if len(best_third_teams) > 7 else None),
        (group_winners.get("B"), best_third_teams[6] if len(best_third_teams) > 6 else None),
        (group_winners.get("C"), best_third_teams[5] if len(best_third_teams) > 5 else None),
        (group_winners.get("D"), best_third_teams[4] if len(best_third_teams) > 4 else None),
        (group_winners.get("E"), best_third_teams[3] if len(best_third_teams) > 3 else None),
        (group_winners.get("F"), best_third_teams[2] if len(best_third_teams) > 2 else None),
        (group_winners.get("G"), best_third_teams[1] if len(best_third_teams) > 1 else None),
        (group_winners.get("H"), best_third_teams[0] if len(best_third_teams) > 0 else None),
        (group_runners.get("A"), group_runners.get("B")),
        (group_runners.get("C"), group_runners.get("D")),
        (group_runners.get("E"), group_runners.get("F")),
        (group_runners.get("G"), group_runners.get("H")),
        (group_winners.get("I"), group_runners.get("J")),
        (group_winners.get("K"), group_runners.get("L")),
        (group_winners.get("J"), group_runners.get("I")),
        (group_winners.get("L"), group_runners.get("K"))
    ]

    r32_teams = []
    round16 = []
    for team1, team2 in round_of_32:
        r32_teams.extend([team1, team2])
        round16.append(simulate_knockout(team1, team2, "Round of 32", all_matches, deterministic))

    quarterfinals = play_round(round16, "Round of 16", all_matches, deterministic)
    semifinals = play_round(quarterfinals, "Quarterfinals", all_matches, deterministic)
    finalists = play_round(semifinals, "Semifinals", all_matches, deterministic)
    champion = simulate_knockout(finalists[0], finalists[1], "Final", all_matches, deterministic)

    # STEP 9: Return everything required for the dashboard
    return {
        "Champion": champion, 
        "Matches": all_matches, 
        "GroupTables": all_group_tables, 
        "Round32": [t for t in r32_teams if t is not None],
        "Round16": round16, 
        "Quarterfinals": quarterfinals, 
        "Semifinals": semifinals, 
        "Finalists": finalists
    }
```
