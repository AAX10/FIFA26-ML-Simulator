```python
# Cell 2: Load Raw Data & Aggregate Player Stats

import glob
import os
import pandas as pd

# =====================================================
# TEAM NAME STANDARDIZATION
# =====================================================

TEAM_MAP = {

    "Korea Republic": "South Korea",

    "IR Iran": "Iran",

    "USA": "United States",

    "Côte d'Ivoire": "Ivory Coast",

    "Congo DR": "DR Congo",

    "Cape Verde Islands": "Cape Verde",

    "Cabo Verde": "Cape Verde",

    "Curacao": "Curaçao",

    "Czechia": "Czech Republic"
}

# =====================================================
# 1. HISTORICAL MATCHES
# =====================================================

raw_matches = pd.read_csv(
    "../data/international_matches/results.csv"
)

raw_matches = raw_matches.rename(columns={
    "date": "Date",
    "home_team": "HomeTeam",
    "away_team": "AwayTeam",
    "home_score": "HomeGoals",
    "away_score": "AwayGoals"
})

raw_matches["Date"] = pd.to_datetime(
    raw_matches["Date"]
)

raw_matches["HomeTeam"] = (
    raw_matches["HomeTeam"]
    .replace(TEAM_MAP)
)

raw_matches["AwayTeam"] = (
    raw_matches["AwayTeam"]
    .replace(TEAM_MAP)
)

# =====================================================
# 2. WORLD CUP HISTORY
# =====================================================

wc_raw = pd.read_csv(
    "../data/historical_world_cups/WorldCups.csv"
)

wc_raw["Winner"] = (
    wc_raw["Winner"]
    .replace(TEAM_MAP)
)

wc_history_df = (
    wc_raw["Winner"]
    .value_counts()
    .reset_index()
)

wc_history_df.columns = [
    "Team",
    "WC_Titles"
]

# =====================================================
# 3. FIFA RANKINGS
# =====================================================

print("Loading FIFA Rankings...")

rankings_df = pd.read_csv(
    "../data/international_matches/fifa_ranking.csv"
)

rankings_df["Date"] = pd.to_datetime(
    rankings_df["rank_date"]
)

rankings_df = rankings_df.rename(columns={
    "country_full": "Team",
    "rank": "Rank",
    "total_points": "Points"
})

rankings_df["Team"] = (
    rankings_df["Team"]
    .replace(TEAM_MAP)
)

# =====================================================
# 4. FIFA PLAYER DATABASE
# =====================================================

print(
    "Aggregating player files... this might take a minute."
)

player_files = glob.glob(
    "../data/fifa_ratings/players_*.csv"
)

squad_list = []

for file in player_files:

    year_str = (
        os.path.basename(file)
        .replace("players_", "")
        .replace(".csv", "")
    )

    if not year_str.isdigit():
        continue

    year = int("20" + year_str)

    df_players = pd.read_csv(
        file,
        low_memory=False
    )

    nat_col = (
        "nationality_name"
        if "nationality_name" in df_players.columns
        else "nationality"
    )

    val_col = (
        "value_eur"
        if "value_eur" in df_players.columns
        else "value"
    )

    df_players = df_players.sort_values(
        "overall",
        ascending=False
    )

    def get_squad_metrics(group):

        top_11 = group.head(11)

        top_23 = group.head(23)

        return pd.Series({

            "Avg_OVR_Top11":
                top_11["overall"].mean(),

            "Total_Value_Top23":
                top_23[val_col].sum()
                if val_col in group.columns
                else 0,

            "Avg_Age":
                top_23["age"].mean()

        })

    squad_metrics = (
        df_players
        .groupby(nat_col)
        .apply(get_squad_metrics)
        .reset_index()
    )

    squad_metrics["Date"] = pd.to_datetime(
        f"{year}-01-01"
    )

    squad_metrics = squad_metrics.rename(
        columns={nat_col: "Team"}
    )

    squad_metrics["Team"] = (
        squad_metrics["Team"]
        .replace(TEAM_MAP)
    )

    squad_list.append(
        squad_metrics
    )
    
# =====================================================
# COMBINE ALL PLAYER FILES
# =====================================================

if len(squad_list) > 0:

    squad_stats_df = pd.concat(
        squad_list,
        ignore_index=True
    )

else:

    raise ValueError(
        "No players_*.csv files found!"
    )

# =====================================================
# DIAGNOSTIC CHECK
# =====================================================

print("\nMapped Teams Check:\n")

for team in [
    "South Korea",
    "Iran",
    "United States",
    "Ivory Coast",
    "DR Congo",
    "Cape Verde",
    "Czech Republic"
]:
    print(
        team,
        "| Rankings:",
        team in rankings_df["Team"].values,
        "| Squad:",
        team in squad_stats_df["Team"].values
    )

print("\nData loading complete!")
```

```python
from src.ranking_system import EloSystem
from src.feature_engineering import compute_rolling_form, extract_form_ledger

# ====================================================================
# --- 1. Generate Dynamic Elo (Using all matches) ---
# ====================================================================
elo_engine = EloSystem()
elo_ledger = elo_engine.process_matches(raw_matches)
elo_ledger['Date'] = pd.to_datetime(elo_ledger['Date'])

# ====================================================================
# --- 2. Generate Pure World Cup Form (Using clean_fifa dataset) ---
# ====================================================================
# Load the clean World Cup dataset
wc_form_df = pd.read_csv('/data/fifa_worldcup_matches.csv')

wc_form_df = wc_form_df.rename(columns={
    "HomeTeam": "HomeTeam", 
    "AwayTeam": "AwayTeam", 
    "HomeGoals": "HomeGoals", 
    "AwayGoals": "AwayGoals"
})

# Proxy Date: Assign Dec 31st of the match Year + 1 minute per row to guarantee strict chronological order!
wc_form_df['Date'] = pd.to_datetime(wc_form_df['Year'].astype(str) + '-12-31') + pd.to_timedelta(wc_form_df.index, unit='m')

# Calculate rolling form STRICTLY on World Cup history
wc_with_form = compute_rolling_form(wc_form_df)

# Extract the unified ledger (Team, Date, Form5, Momentum, etc.)
wc_ledger = extract_form_ledger(wc_with_form)

# ====================================================================
# --- 3. Merge "Bottom Values" Backward onto the Main Timeline ---
# ====================================================================
matches_with_form = raw_matches.copy()
matches_with_form['Date'] = pd.to_datetime(matches_with_form['Date'])
matches_with_form = matches_with_form.sort_values('Date')

# By using direction='backward', any match in the future automatically 
# reaches back and takes the final form values from the team's last WC appearance!
matches_with_form = pd.merge_asof(
    matches_with_form, wc_ledger.add_prefix('Home_'),
    left_on='Date', right_on='Home_Date',
    left_by='HomeTeam', right_by='Home_Team', direction='backward'
).drop(columns=['Home_Date', 'Home_Team'])

matches_with_form = pd.merge_asof(
    matches_with_form, wc_ledger.add_prefix('Away_'),
    left_on='Date', right_on='Away_Date',
    left_by='AwayTeam', right_by='Away_Team', direction='backward'
).drop(columns=['Away_Date', 'Away_Team'])

# Fill defaults for teams before they play their first ever World Cup match
form_fill_cols = [col for col in matches_with_form.columns if "Form" in col or "Momentum" in col or "Pts" in col]
matches_with_form[form_fill_cols] = matches_with_form[form_fill_cols].fillna(1.0)

print("Pipeline Stage Complete: Elo and Pure World Cup Form successfully mapped!")
```

```python
import pandas as pd

shootouts = pd.read_csv(
    "../data/international_matches/shootouts.csv"
)

teams = sorted(
    set(shootouts["home_team"])
    |
    set(shootouts["away_team"])
)

penalty_stats = []

for team in teams:

    total = (
        (shootouts["home_team"] == team).sum()
        +
        (shootouts["away_team"] == team).sum()
    )

    wins = (
        shootouts["winner"] == team
    ).sum()

    losses = total - wins

    rating = (
        wins / total
        if total > 0
        else 0.5
    )

    penalty_stats.append({

        "Team": team,

        "Penalty_Wins": wins,

        "Penalty_Losses": losses,

        "Penalty_Rating": rating

    })

penalty_df = pd.DataFrame(
    penalty_stats
)

print(penalty_df.head())
print("Teams:", len(penalty_df))
```

```python
# Execute the comprehensive merge to combine all 5 datasets
final_df = build_comprehensive_features(

    matches=matches_with_form,

    elo_ledger=elo_ledger,

    rankings=rankings_df,

    squad_stats=squad_stats_df,

    wc_history=wc_history_df,

    wc_profiles=wc_profiles,

    penalty_df=penalty_df

)
teams = set(final_df["HomeTeam"]).union(
    set(final_df["AwayTeam"])
)

print("Total teams:", len(teams))

for t in sorted(teams):
    if (
        "Korea" in t
        or "Iran" in t
        or "United" in t
        or "USA" in t
        or "Ivoire" in t
        or "Ivory" in t
        or "Congo" in t
    ):
        print(t)
# Export the ML-ready dataset
final_df.to_csv('../data/world_cup_prediction_dataset/train.csv', index=False)
print(f"Pipeline complete! Final dataset shape: {final_df.shape}")
```
