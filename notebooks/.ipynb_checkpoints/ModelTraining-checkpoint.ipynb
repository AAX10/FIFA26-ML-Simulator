```python
%load_ext autoreload
%autoreload 2

import os
import sys
import joblib
import pandas as pd
import numpy as np

sys.path.append(os.path.abspath(".."))

from src.match_predictor import train_time_series_models
```

```python
df = pd.read_csv(
    "../data/world_cup_prediction_dataset/train.csv"
)

df["Date"] = pd.to_datetime(df["Date"])

print("Rows:", len(df))
print("Columns:", len(df.columns))

display(df.head())
```

```python
if "Target" not in df.columns:

    # Home Win = 2
    # Draw = 1
    # Away Win = 0

    df["Target"] = 0

    df.loc[
        df["HomeGoals"] == df["AwayGoals"],
        "Target"
    ] = 1

    df.loc[
        df["HomeGoals"] > df["AwayGoals"],
        "Target"
    ] = 2

print(df["Target"].value_counts())
```

```python
required_features = [

    "Elo_Delta",

    "Rank_Delta",

    "Points_Delta",

    "OVR_11_Delta",

    "Value_Delta",

    "WC_Pedigree_Delta",

    "Home_Form_Pts",

    "Away_Form_Pts",

    "Form_Pts_Delta",

    "Form_GS_Delta",

    "Form_GC_Delta",

    "Goal_Diff_Form",

    "Penalty_Delta",

    "Weighted_Form_Delta",

    "Form5_Delta",

    "Form10_Delta",

    "Form20_Delta",

    "Momentum_Delta"
]

missing = [
    col
    for col in required_features
    if col not in df.columns
]

print("Missing Features:")
print(missing)
```

```python
usable_features = [
    c
    for c in required_features
    if c in df.columns
]

training_cols = (
    usable_features
    + [
        "Target",
        "HomeGoals",
        "AwayGoals"
    ]
)
fill_values = {

    "Home_Avg_OVR_Top11": 75,
    "Away_Avg_OVR_Top11": 75,

    "Home_Total_Value_Top23":
        df["Home_Total_Value_Top23"].median(),

    "Away_Total_Value_Top23":
        df["Away_Total_Value_Top23"].median(),

    "Home_Elo":
        df["Home_Elo"].median(),

    "Away_Elo":
        df["Away_Elo"].median(),

    "Home_Rank":
        df["Home_Rank"].median(),

    "Away_Rank":
        df["Away_Rank"].median(),

    "Home_Points":
        df["Home_Points"].median(),

    "Away_Points":
        df["Away_Points"].median()
}

df = df.fillna(fill_values)
df = df.dropna(
    subset=[
        "Target",
        "HomeGoals",
        "AwayGoals"
    ]
)

print("Training Rows:", len(df))
```

```python
print(df[usable_features].describe())
```

```python
clf, reg_home, reg_away = (
    train_time_series_models(df)
)
```

```python
print("\nMODEL FEATURES:")
print(clf.feature_names_in_)
print("\nTOTAL FEATURES:")
print(len(clf.feature_names_in_))
```

```python
os.makedirs(
    "../models",
    exist_ok=True
)

joblib.dump(
    clf,
    "../models/classifier.pkl"
)

joblib.dump(
    reg_home,
    "../models/regressor_home.pkl"
)

joblib.dump(
    reg_away,
    "../models/regressor_away.pkl"
)

print("Models saved.")

model = joblib.load(
    "../models/classifier.pkl"
)

print(model.feature_names_in_)
print(len(model.feature_names_in_))
```

```python
import matplotlib.pyplot as plt

importance = pd.DataFrame(
    {
        "Feature": clf.feature_names_in_,
        "Importance": clf.feature_importances_
    }
)

importance = (
    importance
    .sort_values(
        "Importance",
        ascending=False
    )
)

display(importance)

plt.figure(figsize=(10,6))

plt.barh(
    importance["Feature"],
    importance["Importance"]
)

plt.tight_layout()
plt.show()
```

```python
sample = df.iloc[-1:].copy()

model_features = clf.feature_names_in_

X_sample = sample[model_features]

probs = clf.predict_proba(X_sample)[0]

print(f"Away Win : {probs[0]:.3f}")
print(f"Draw     : {probs[1]:.3f}")
print(f"Home Win : {probs[2]:.3f}")

home_goals = reg_home.predict(X_sample)[0]
away_goals = reg_away.predict(X_sample)[0]

print(
    f"\nPredicted Score: "
    f"{round(home_goals)} - {round(away_goals)}"
)
```
