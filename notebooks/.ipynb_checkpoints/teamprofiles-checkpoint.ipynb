```python
import pandas as pd

# Load final training dataset
df = pd.read_csv(
    "../data/world_cup_prediction_dataset/train.csv"
)

df["Date"] = pd.to_datetime(df["Date"])

# Most recent matches first
df = df.sort_values(
    "Date",
    ascending=False
)

team_profiles = {}

for _, row in df.iterrows():

    home = row["HomeTeam"]
    away = row["AwayTeam"]

    # Save latest HomeTeam stats
    if home not in team_profiles:

        team_profiles[home] = {

            "Team": home,
        
            "Elo": row["Home_Elo"],
        
            "Rank": row["Home_Rank"],
        
            "Points": row["Home_Points"],
        
            "OVR11": row["Home_Avg_OVR_Top11"],
        
            "Value": row["Home_Total_Value_Top23"],
        
            "WC_Titles": row["Home_WC_Titles"],
        
            "Form_Pts": row["Home_Form_Pts"],
        
            "Form_GS": row["Home_Form_GS"],
        
            "Form_GC": row["Home_Form_GC"]
        }

    # Save latest AwayTeam stats
    if away not in team_profiles:

        team_profiles[away] = {

            "Team": away,
        
            "Elo": row["Away_Elo"],
        
            "Rank": row["Away_Rank"],
        
            "Points": row["Away_Points"],
        
            "OVR11": row["Away_Avg_OVR_Top11"],
        
            "Value": row["Away_Total_Value_Top23"],
        
            "WC_Titles": row["Away_WC_Titles"],
        
            "Form_Pts": row["Away_Form_Pts"],
        
            "Form_GS": row["Away_Form_GS"],
        
            "Form_GC": row["Away_Form_GC"]
        }

profiles_df = pd.DataFrame(
    team_profiles.values()
)

profiles_df = profiles_df.sort_values(
    "Team"
)

profiles_df.to_csv(
    "../data/team_profiles.csv",
    index=False
)

print(f"Created {len(profiles_df)} teams")

display(profiles_df.head())
```
