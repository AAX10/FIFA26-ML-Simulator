```python
import pandas as pd

fixtures_df = pd.read_csv(
    "../data/world_cup_2026_fixtures.csv"
)

fixtures_df["Group"] = (
    fixtures_df["Group"]
    .astype(str)
    .str.strip()
)

fixtures_df["Home"] = (
    fixtures_df["Home"]
    .astype(str)
    .str.strip()
)

fixtures_df["Away"] = (
    fixtures_df["Away"]
    .astype(str)
    .str.strip()
)

groups = pd.concat([

    fixtures_df[
        ["Group","Home"]
    ].rename(
        columns={
            "Home":"Team"
        }
    ),

    fixtures_df[
        ["Group","Away"]
    ].rename(
        columns={
            "Away":"Team"
        }
    )

])

groups = (
    groups
    .drop_duplicates()
    .sort_values(
        ["Group","Team"]
    )
)

print(
    "Total Teams:",
    len(groups)
)
```

```python
import requests

API_URL = "http://127.0.0.1:8000"

def predict_match(home, away):

    payload = {

        "home_team": home,

        "away_team": away,

        "is_neutral": True

    }

    response = requests.post(
        f"{API_URL}/predict",
        json=payload
    )

    if response.status_code != 200:

        print(
            f"FAILED: {home} vs {away}"
        )
    
        print(
            response.status_code
        )
    
        print(
            response.text
        )
    
        return None

    return response.json()
```

```python
import pandas as pd
import random

def simulate_group(
    group,
    teams,
    verbose=False
):

    table = pd.DataFrame(
        index=teams,
        columns=[
            "P",
            "W",
            "D",
            "L",
            "Pts",
            "GF",
            "GA",
            "GD"
        ]
    ).fillna(0)

    group_fixtures = (
        fixtures_df[
            fixtures_df["Group"] == group
        ]
        .sort_values("Matchday")
    )

    for _, match in group_fixtures.iterrows():

        home = match["Home"]
        away = match["Away"]

        result = predict_match(
            home,
            away
        )

        if result is None:
            continue

        probs = result["probabilities"]

        outcome = random.choices(

            ["H","D","A"],

            weights=[
                probs["home_win"],
                probs["draw"],
                probs["away_win"]
            ]

        )[0]

        hg = result["predicted_score"]["home_goals"]
        ag = result["predicted_score"]["away_goals"]

        table.loc[home,"P"] += 1
        table.loc[away,"P"] += 1

        table.loc[home,"GF"] += hg
        table.loc[home,"GA"] += ag

        table.loc[away,"GF"] += ag
        table.loc[away,"GA"] += hg

        if outcome == "H":

            table.loc[home,"Pts"] += 3

            table.loc[home,"W"] += 1

            table.loc[away,"L"] += 1

        elif outcome == "A":

            table.loc[away,"Pts"] += 3

            table.loc[away,"W"] += 1

            table.loc[home,"L"] += 1

        else:

            table.loc[home,"Pts"] += 1
            table.loc[away,"Pts"] += 1

            table.loc[home,"D"] += 1
            table.loc[away,"D"] += 1

    table["GD"] = (
        table["GF"]
        - table["GA"]
    )

    table = table.sort_values(

        ["Pts","GD","GF"],

        ascending=False

    )

    return table
```

```python
import random

def simulate_knockout(team1, team2):

    result = predict_match(
        team1,
        team2
    )

    if result is None:

        return team1

    probs = result["probabilities"]

    outcome = random.choices(

        ["H", "D", "A"],

        weights=[

            probs["home_win"],

            probs["draw"],

            probs["away_win"]

        ]

    )[0]

    if outcome == "H":

        return team1

    elif outcome == "A":

        return team2

    else:

        return result[
            "knockout_prediction"
        ]["winner"]
```

```python
def play_round(teams):

    winners = []

    for i in range(
        0,
        len(teams),
        2
    ):

        winner = simulate_knockout(

            teams[i],

            teams[i + 1]

        )

        winners.append(
            winner
        )

    return winners
```

```python
def simulate_world_cup():

    group_winners = {}
    group_runners = {}

    all_group_tables = {}

    for group in sorted(
        groups["Group"].unique()
    ):

        teams = groups[
            groups["Group"] == group
        ]["Team"].tolist()

        table = simulate_group(
            group,
            teams,
            verbose=False
        )

        all_group_tables[group] = table

        group_winners[group] = table.index[0]

        group_runners[group] = table.index[1]

    third_place = []

    for group, table in all_group_tables.items():

        third_place.append({

            "Group": group,

            "Team": table.index[2],

            "Pts": table.iloc[2]["Pts"],

            "GD": table.iloc[2]["GD"],

            "GF": table.iloc[2]["GF"]

        })

    third_place = pd.DataFrame(
        third_place
    )

    best_8 = third_place.sort_values(

        ["Pts","GD","GF"],

        ascending=False

    ).head(8)

    best_third_teams = (
        best_8["Team"]
        .tolist()
    )

    round_of_32 = [

        (group_winners["A"], best_third_teams[7]),
        (group_winners["B"], best_third_teams[6]),
        (group_winners["C"], best_third_teams[5]),
        (group_winners["D"], best_third_teams[4]),

        (group_winners["E"], best_third_teams[3]),
        (group_winners["F"], best_third_teams[2]),
        (group_winners["G"], best_third_teams[1]),
        (group_winners["H"], best_third_teams[0]),

        (group_runners["A"], group_runners["B"]),
        (group_runners["C"], group_runners["D"]),
        (group_runners["E"], group_runners["F"]),
        (group_runners["G"], group_runners["H"]),

        (group_winners["I"], group_runners["J"]),
        (group_winners["K"], group_runners["L"]),
        (group_winners["J"], group_runners["I"]),
        (group_winners["L"], group_runners["K"])

    ]

    round16 = []

    for team1, team2 in round_of_32:

        round16.append(

            simulate_knockout(
                team1,
                team2
            )

        )

    quarterfinals = play_round(
        round16
    )

    semifinals = play_round(
        quarterfinals
    )

    finalists = play_round(
        semifinals
    )

    champion = simulate_knockout(

        finalists[0],

        finalists[1]

    )

    return champion
```

```python
from collections import Counter

N_SIMS = 5000

champions = Counter()

for i in range(N_SIMS):

    champion = simulate_world_cup()

    champions[champion] += 1

print(
    "\nWORLD CUP WIN PROBABILITIES\n"
)

for team, count in champions.most_common():

    print(

        f"{team:<20}"

        f"{100*count/N_SIMS:.2f}%"

    )
```
