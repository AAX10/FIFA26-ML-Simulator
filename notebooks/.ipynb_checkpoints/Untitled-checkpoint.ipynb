```python
import pandas as pd

shootouts = pd.read_csv(
    "../data/international_matches/shootouts.csv"
)

wins = {}
losses = {}

for _, row in shootouts.iterrows():

    winner = row["winner"]

    home = row["home_team"]
    away = row["away_team"]

    loser = away if winner == home else home

    wins[winner] = wins.get(winner, 0) + 1
    losses[loser] = losses.get(loser, 0) + 1

teams = set(list(wins.keys()) + list(losses.keys()))

penalty_stats = []

for team in teams:

    w = wins.get(team, 0)
    l = losses.get(team, 0)

    total = w + l

    if total == 0:
        rate = 0.5
    else:
        rate = w / total

    penalty_stats.append({
        "Team": team,
        "Penalty_Wins": w,
        "Penalty_Losses": l,
        "Penalty_Rating": rate
    })

penalty_df = pd.DataFrame(penalty_stats)

display(
    penalty_df.sort_values(
        "Penalty_Rating",
        ascending=False
    ).head(20)
)
```
